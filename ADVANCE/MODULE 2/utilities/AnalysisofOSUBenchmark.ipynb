```python
import pandas as pd
from typing import Dict
import matplotlib.pyplot as plt

from ConverterTxtToPd import load_all_benchmark_results, clean_benchmark_df

# Set the path
directory_path = "C:/Users/matte/Documents/CLOUD/CloudProject/NextCloudDeployment/ADVANCE/MODULE 2/store"
```

```python
results = load_all_benchmark_results(directory_path)
```

```python
type(results['allreduce-one-node-results.txt'])
```

```python
df_p2p_2 = results['p2p-two-nodes-results.txt']
df_p2p_2 = df_p2p_2.rename(columns={
    'Size          Latency': 'Size'
})
df_p2p_1 = results['p2p-one-node-results.txt']
df_p2p_1 = df_p2p_1.rename(columns={
    'Size          Latency': 'Size'
})
df_bc_2 = clean_benchmark_df(results['bcast-two-nodes-results.txt'])
df_bc_1 = clean_benchmark_df(results['bcast-one-node-results.txt'])
df_all_2 = clean_benchmark_df(results['allreduce-two-nodes-results.txt'])
df_all_1 = clean_benchmark_df(results['allreduce-one-node-results.txt'])
```

# LATENCY

```python
plt.clf()
plt.figure(figsize=(10,10))

# One Node
plt.semilogx(df_p2p_1['Size'], df_p2p_1['us'], 
             marker="o", color="blue", label="One Node", linewidth=2, markersize=5)

# Two Nodes
plt.semilogx(df_p2p_2['Size'], df_p2p_2['us'], 
             marker="s", color="red", label="Two Nodes", linewidth=2, markersize=5)

plt.xlabel("Message Size (bytes)")
plt.ylabel("Latency (s)")
plt.title("Benchmark - Latency")
plt.grid(True, alpha=0.3, which="both")
plt.tight_layout()
plt.legend()
plt.show()
```
